# Muestras derivadas del modelo de entrenamiento con kernels

```python
from matplotlib import pyplot as plt
from plotting import get_comparatives, plot_comparatives, view_plots_and_save_them

n_component_column = 'n-component'
std_column_general_name = 'std_column'

column_name_list = ['training-average-time', 'predict-average-time']
names_of_std_for_column = ['score-std-time', 'predict-std-time']                   


for index,column_name in enumerate(column_name_list):
    transform = ['RFF', 'Nystroem']
    models = ['ridge', 'svc', 'linear-svc']

    list_of_column = [f'{t}-{m}-{column_name}' for t in transform  for m in models]
    list_of_std_column = [f'{t}-{m}-{names_of_std_for_column[index]}' for t in transform  for m in models]
    list_of_models = [f'{t}-{m}' for t in transform  for m in models]

    df_with_n_component = []
    constants_lines = []
    constant_names = ['kernel-svc', 'kernel-ridge']
    constants_std = []
    # Sacamos las constantes de las desviaciones típicas
    for  kernel,std in zip(constant_names,names_of_std_for_column):
        constants_lines.append(kernel_df[f'{kernel}-{column_name}'].dropna().to_list()[0])
        if column_name_list[0] == column_name:
            constants_std.append(0)
        else:
            constants_std.append(kernel_df[f'{kernel}-{std}'].dropna().to_list()[0])

    names_list = []
    for index,column in enumerate(list_of_column):
        std_column_name = list_of_std_column[index]
        print([n_component_column,column, std_column_name])
        df_with_n_component.append(pd.DataFrame(random_features_df[[n_component_column,column, std_column_name]]))
        df_with_n_component[-1].rename(columns={column: column_name, std_column_name: std_column_general_name}, inplace=True)
    for bigger_than in [0, bigger_than_database]:
        title = column_name + f'-bigger-than-{bigger_than}-components-with-kernels-with-std'
        title += '-'+dataset
        get_comparatives(column=column_name, df_with_n_components=df_with_n_component, 
                        df_labels=list_of_models, title= title,
                        column_x = n_component_column, axis_x_name = 'n_components',
                        names_of_std_for_column={column_name:std_column_general_name},
                        percent_of_n_components_bigger_than=bigger_than,
                        constant_labels=constant_names, constant_lines=constants_lines, constant_std=constants_std)
        
        save_route =  plot_path + title
        # Guarda el gráfico en la ubicación especificada por save_path
        plt.savefig(save_route, bbox_inches='tight')  # bbox_inches='tight' ajusta los márgenes para que se ajusten correctamente
        
        plt.show()




```
